```python
import findspark
findspark.init("/Users/chukuemekaogudu/Documents/Dev-Spark-Apache/Apache-Spark/spark-2.4.5-bin-hadoop2.7")

import os
import json
from pyspark import SparkContext, SparkConf
```

```python
data_dir = "/Volumes/oli2/inf533_datasets/"
```

```python
config = SparkConf().setMaster("local[*]") \
                  .set("spark.executor.memory", "4g") \
                  .set("spark.driver.memory", "4g")
sc = SparkContext(conf=config).getOrCreate()
sc.setLogLevel("ERROR")
```

```python
import binascii
```

```python
first = sc.textFile(os.path.join(data_dir, "business_first.json")) \
          .map(json.loads) \
          .map(lambda x: x.get("city")) \
          .filter(lambda x: x != None and x != "") \
          .distinct() \
          .map(lambda x: int(binascii.hexlify(x.encode("utf8")), 16))
```

```python
import random
random.seed(25)
```

```python
def hash_funcs():
    p = 2**35 - 355
    a = random.randint(1, p - 1)
    b = random.randint(0, p - 1)
    m = 8351
    return lambda x: ((a * x + b) % p) % m
```

```python
num_hash = 7
```

```python
bit_vector = [0] * 8351
```

```python
hash_list = [hash_funcs() for _ in range(num_hash)]
```

```python
hash_pos = first.map(lambda x: [func(x) for func in hash_list]) \
                .flatMap(lambda x: x) \
                .distinct() \
                .collect()
```

```python
for pos in hash_pos:
    bit_vector[pos] = 1
```

```python
second = sc.textFile(os.path.join(data_dir, "business_second.json")) \
           .map(json.loads) \
           .map(lambda x: x.get("city")) \
           .collect()
```

```python
predictions = []
```

```python
for city in second:
    if city != None and city != "":
        city = int(binascii.hexlify(city.encode("utf8")), 16)
        bit_val = [bit_vector[func(city)] for func in hash_list]
        if sum(bit_val) == len(bit_val):
            predictions.append("1")
        else:
            predictions.append("0")
    else:
        predictions.append("0")
```
